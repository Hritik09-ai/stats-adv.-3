```python
# QUES. 1 What is Estimation Statistics? Explain point estimate and interval estimate.
# ANSWER  
# Estimation statistics is a branch of statistics concerned with estimating population parameters based on sample data.
# When it's impractical or impossible to collect data from an entire population, statisticians use estimation techniques
# to infer characteristics about the population based on information gathered from a representative sample.

# There are two main types of estimation in statistics: point estimation and interval estimation.

# 1. Point Estimate:

# * A point estimate is a single value that serves as the best guess or approximation of the population parameter. It's 
# calculated directly from the sample data and represents the most likely value of the parameter.
# * For example, if you're estimating the population mean based on a sample mean, the sample mean would be the point 
# estimate of the population mean.

# 2. Interval Estimate:

# * An interval estimate provides a range of values within which the true population parameter is likely to fall, along with
# a level of confidence associated with that range.
# * Interval estimates are often expressed in the form of confidence intervals. A confidence interval consists of a lower
# bound and an upper bound, between which the population parameter is estimated to lie with a specified level of confidence.
# * For example, a 95% confidence interval for the population mean is a range of values within which we are 95% confident that
# the true population mean lies.
```

```python
# QUES.2 Write a Python function to estimate the population mean using a sample mean and standard
# deviation.

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    """
    Function to estimate the population mean using sample mean and standard deviation.

    Parameters:
        sample_mean (float): The mean of the sample.
        sample_std_dev (float): The standard deviation of the sample.
        sample_size (int): The size of the sample.

    Returns:
        float: Estimated population mean.
    """
    import math
    standard_error = sample_std_dev / math.sqrt(sample_size)
    population_mean = sample_mean * (1 / standard_error)
    return population_mean

# Example usage:
sample_mean = 50.5
sample_std_dev = 10.2
sample_size = 100
estimated_population_mean = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print("Estimated Population Mean:", estimated_population_mean)
```

```python
# QUES. 3 What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
# ANSWER Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It 
# involves formulating two competing hypotheses: the null hypothesis (H0) and the alternative hypothesis (H1). The null 
# hypothesis typically represents a default assumption or the status quo, while the alternative hypothesis suggests a 
# deviation from the null hypothesis, such as a difference or an effect.

# The process of hypothesis testing involves collecting sample data and using statistical techniques to determine whether 
# there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis. This is done by calculating
# a test statistic based on the sample data and comparing it to a critical value or using a p-value to determine the 
# likelihood of observing the data under the null hypothesis.

# Hypothesis testing is used for several reasons:

# 1. Making Inferences: It allows researchers to make inferences about population parameters based on sample data. By testing
# hypotheses, researchers can draw conclusions about the population from which the sample was drawn.

# 2. Decision Making: Hypothesis testing provides a formal framework for decision making. It helps decision-makers evaluate
# the significance of results and determine whether to take action based on those results.

# 3. Comparing Groups: It enables comparison between two or more groups or treatments to determine if there is a significant
# difference between them. This is particularly useful in experimental research where different conditions are applied to
# groups to assess their effects.

# 4. Control and Quality Assurance: Hypothesis testing is used in quality control processes to ensure that products or 
# processes meet specified standards. By testing hypotheses about the quality or performance of products or processes,
# organizations can identify areas for improvement and maintain consistency.

# 5. Scientific Research: In scientific research, hypothesis testing is fundamental for testing theories and hypotheses 
# about natural phenomena. It helps scientists evaluate the validity of their hypotheses and contributes to the
# accumulation of knowledge in various fields.
```

```python
# QUES. 5 Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.
# AMSWER 
import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05, alternative='two-sided'):
    """
    Perform two-sample t-test for the difference of means.

    Parameters:
    sample1 : array_like
        The first sample data.
    sample2 : array_like
        The second sample data.
    alpha : float, optional
        The significance level for the test (default is 0.05).
    alternative : {'two-sided', 'less', 'greater'}, optional
        Specifies the alternative hypothesis.
        - 'two-sided' (default): the difference in means is not equal to the null hypothesis value.
        - 'less': the difference in means is less than the null hypothesis value.
        - 'greater': the difference in means is greater than the null hypothesis value.

    Returns:
    t_statistic : float
        The calculated t-statistic.
    p_value : float
        The two-tailed p-value.
    """

    # Calculate sample statistics
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    n1 = len(sample1)
    n2 = len(sample2)
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate pooled standard deviation
    pooled_std = np.sqrt(var1 / n1 + var2 / n2)

    # Calculate t-statistic
    t_statistic = (mean1 - mean2) / pooled_std

    # Calculate degrees of freedom
    df = (var1 / n1 + var2 / n2)**2 / ((var1**2 / (n1**2 * (n1 - 1))) + (var2**2 / (n2**2 * (n2 - 1))))

    # Calculate p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))
    elif alternative == 'less':
        p_value = stats.t.cdf(t_statistic, df)
    elif alternative == 'greater':
        p_value = 1 - stats.t.cdf(t_statistic, df)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'.")

    return t_statistic, p_value

# Example usage:
sample1 = np.random.normal(loc=5, scale=2, size=30)  # Sample from population 1
sample2 = np.random.normal(loc=6, scale=2, size=30)  # Sample from population 2

t_statistic, p_value = two_sample_t_test(sample1, sample2)

print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# QUES. 6 What is a null and alternative hypothesis? Give some examples.
# ANSWER In statistical hypothesis testing, the null hypothesis (denoted as H0) and the alternative hypothesis (denoted
# as Ha or H1) are two competing statements about the population parameters. These hypotheses are used to assess the 
# validity of a claim or to determine if there is enough evidence to support a particular assertion.

# 1. Example 1 - Drug Efficacy:

# * Null Hypothesis (H0): The mean effectiveness of a new drug is equal to that of the existing drug.
# * Alternative Hypothesis (Ha): The mean effectiveness of the new drug is different from that of the existing drug.
# * Here, the null hypothesis suggests no difference in effectiveness, while the alternative hypothesis suggests a difference.

# 2. Example 2 - Coin Toss:

# * Null Hypothesis (H0): The coin is fair (i.e., the probability of getting heads is 0.5).
# * Alternative Hypothesis (Ha): The coin is not fair (i.e., the probability of getting heads is not equal to 0.5).
# * In this case, the null hypothesis assumes fairness, while the alternative hypothesis considers the possibility of 
# unfairness.

# 3. Example 3 - Advertising Impact:

# * Null Hypothesis (H0): The mean sales before and after an advertising campaign are equal.
# * Alternative Hypothesis (Ha): The mean sales after the advertising campaign are higher than before.
# * The null hypothesis assumes no change in sales, while the alternative hypothesis suggests an increase after the 
# advertising campaign.

# 4. Example 4 - Educational Intervention:

# * Null Hypothesis (H0): The average test scores of students who received an educational intervention are the same as those
# who did not.
# * Alternative Hypothesis (Ha): The average test scores of students who received an educational intervention are higher than
# those who did not.
# * Here, the null hypothesis implies no effect of the intervention, while the alternative hypothesis suggests a positive
# effect.
```

```python
# QUES.7 Write down the steps involved in hypothesis testing.
# ANSWER Hypothesis testing is a statistical method used to make inferences about a population parameter based on sample
# data. Here are the general steps involved in hypothesis testing:

# 1. Formulate Hypotheses:

# * State the null hypothesis (H0): This hypothesis typically represents the status quo or no effect. It states that there is
# no significant difference or relationship between variables.
# * State the alternative hypothesis (H1 or Ha): This hypothesis represents what you are trying to prove or find evidence for.
# It could suggest a difference, effect, or relationship between variables.

# 2. Choose a Significance Level (α):

# * The significance level, denoted by α, is the probability of rejecting the null hypothesis when it is actually true. 
# Commonly used values for α are 0.05 or 0.01, indicating a 5% or 1% chance of Type I error, respectively.

# 3. Select the Appropriate Test Statistic:

# * Choose a statistical test based on the type of data (e.g., t-test, z-test, chi-square test) and the nature of the 
# hypothesis (e.g., one-tailed or two-tailed).

# 4. Collect Data and Calculate Test Statistic:

# * Collect a sample from the population of interest.
# Compute the test statistic based on the sample data and the chosen statistical test.

# 5. Determine the Critical Region:

# * Based on the significance level (α) and the chosen statistical test, determine the critical value(s) or critical 
# region(s) from the corresponding probability distribution (e.g., t-distribution, normal distribution).

# 6. Make a Decision:

# * Compare the calculated test statistic to the critical value(s) or critical region(s).
# * If the test statistic falls within the critical region, reject the null hypothesis. If it falls outside the critical
# region, fail to reject the null hypothesis.

# 7. Draw Conclusion:

# * If the null hypothesis is rejected, conclude that there is sufficient evidence to support the alternative hypothesis.
# * If the null hypothesis is not rejected, conclude that there is not enough evidence to support the alternative hypothesis.

# 8. Interpret Results:

# * Interpret the findings in the context of the problem and draw implications or make recommendations based on the results.

# 9. Check Assumptions and Verify Results:

# * Ensure that the assumptions of the statistical test are met and validate the results obtained from hypothesis testing.

# 10. Report Findings:

# * Clearly report the findings, including the test statistic, critical value(s), p-value (if applicable), decision regarding
# the null hypothesis, and interpretation of results.
```

```python
# QUES.8 Define p-value and explain its significance in hypothesis testing.
# ANSWER In statistics, the p-value is a measure used to determine the significance of an observed result in a hypothesis
# test. It represents the probability of obtaining an observed result, or one more extreme, under the assumption that the 
# null hypothesis is true.

# Here's a breakdown of its significance in hypothesis testing:

# 1. Hypothesis Testing Framework: Hypothesis testing involves making decisions about population parameters based on sample 
# data. Typically, there are two competing hypotheses: the null hypothesis (denoted as H0), which represents the default
# assumption, and the alternative hypothesis (denoted as Ha), which represents the claim one seeks to test.

# 2. Comparison with a Threshold (Significance Level): The p-value is compared to a pre-defined threshold called the 
# significance level (α). Commonly used significance levels include 0.05, 0.01, or 0.10. If the p-value is less than or 
# equal to the significance level, it suggests that the observed result is statistically significant, and we reject the 
# null hypothesis in favor of the alternative hypothesis. If the p-value is greater than the significance level, we fail 
# to reject the null hypothesis.

# 3. Interpretation: A small p-value (typically ≤ α) indicates that the observed result is unlikely to have occurred by random
# chance under the assumption that the null hypothesis is true. Therefore, it provides evidence against the null hypothesis
# and supports the alternative hypothesis. On the other hand, a large p-value suggests that the observed result is not
# statistically significant, providing insufficient evidence to reject the null hypothesis.

# 4. Decision Making: The decision to reject or fail to reject the null hypothesis based on the p-value depends on the chosen
# significance level and the context of the study. It's important to note that failing to reject the null hypothesis does 
# not necessarily prove that the null hypothesis is true; it simply means that there is insufficient evidence to reject it
# at the chosen significance level.

# 5. Limitations: While the p-value is a widely used tool in statistical inference, it has limitations. It does not provide
# information about the size or importance of an effect, the precision of estimates, or the probability of the null
# hypothesis being true. Therefore, it should be interpreted in conjunction with other measures and considerations, such
# as effect size, confidence intervals, and practical significance. Additionally, the p-value is sensitive to sample size,
# so a small p-value does not necessarily imply a large effect size or practical significance if the sample size is large.
```

```python
# QUES.10 Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.
# ANSWER 
import numpy as np
from scipy.stats import t

def independent_t_test(sample1, sample2):
    # Calculate means
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)

    # Calculate variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate sample sizes
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate pooled standard deviation
    pooled_std = np.sqrt((var1 + var2) / 2)

    # Calculate t-statistic
    t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(2 / n1))

    # Degrees of freedom
    degrees_of_freedom = n1 + n2 - 2

    # Calculate p-value
    p_value = 2 * (1 - t.cdf(np.abs(t_statistic), degrees_of_freedom))

    return t_statistic, p_value

# Example usage:
sample1 = np.random.normal(loc=5, scale=2, size=30)
sample2 = np.random.normal(loc=5.5, scale=2, size=30)

t_statistic, p_value = independent_t_test(sample1, sample2)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# QUES. 12 What is t-statistic? State the formula for t-statistic.
# ANSWER The t-statistic is a measure used in hypothesis testing to determine if there is a significant difference between
# the means of two groups. It's particularly useful when the sample size is small and the population standard deviation is 
# unknown.

# The formula for the t-statistic is:
# t = x1 - x2 / sP(1/n1+1/n2).sq
# Where:
# * x1 and x2 are the sample means of the two groups being compared.
# * sP is the pooled standard deviation, calculated as:
#     sP = sqroot((n1-1)(s1).sq + (n2-1)(s2).sq/n1+n2-1)
# here sP1.sq and sP2.sq are the sample variances of the two groups, and n1 and n2 are the sample sizes of the two groups.
# * n1 and n2 are the sample sizes of the two groups.
```
